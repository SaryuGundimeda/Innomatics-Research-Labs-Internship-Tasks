# Running Sum of 1d Array

```python
class Solution:
  def runningSum(self, nums: list[int]) -> list[int]:
    for i in range(1, len(nums)):
      nums[i] = nums[i-1] + nums[i]
    return nums

if __name__ == '__main__':
  sol = Solution()
  input = [1,2,3,4]
  print(sol.runningSum(input))
```

# Shuffle the Array

```python
class Solution:
  def shuffle(self, nums: list[int], n: int) -> list[int]:
    output = []

    for i in range(n):
      output.append(nums[i])
      output.append(nums[i+n])
    return output

if __name__ == '__main__':
  sol = Solution()
  input = [2,5,1,3,4,7]
  n = len(input) // 2
  print(sol.shuffle(input, n))
```

# Kids With the Greatest Number of Candies

```python
class Solution:
  def kidsWithCandies(self, candies: list[int], extraCandies: int) -> list[bool]:
    greatest_candy = max(candies)
    output = []

    for i in candies:
      if i + extraCandies >= greatest_candy:
        output.append(True)
      else:
        output.append(False)
    return output

if __name__ == '__main__':
  sol = Solution()
  input = [2,3,5,1,3]
  extraCandies = 3
  print(sol.kidsWithCandies(input, extraCandies))
```
