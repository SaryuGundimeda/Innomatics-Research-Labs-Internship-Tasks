# Defanging an IP Address

```python
class Solution:
  def defangIPaddr(self, address: str) -> str:
    address = address.replace('.', "[.]")
    return address

if __name__ == '__main__':
  sol = Solution()
  input = "1.1.1.1"
  print(sol.defangIPaddr(input))
```

# Find Numbers with an Even Number of Digits

```python
class Solution:
  def findNumbers(self, nums: list[int]) -> int:
    result = 0
    for i in nums:
      if len(str(i)) % 2 == 0:
        result += 1
    return result

if __name__ == '__main__':
  sol = Solution()
  input = [12,345,2,6,7896]
  print(sol.findNumbers(input))
```

# Number of Good Pairs

```python
class Solution:
  def numIdenticalPairs(self, nums: list[int]) -> int:
    count = 0
    for i in range(len(nums)):
      for j in range(i+1,len(nums)):
        if nums[i] == nums[j] and i < j:
          count += 1
    return count

if __name__ == '__main__':
  sol = Solution()
  input = [1,2,3,1,1,3]
  print(sol.numIdenticalPairs(input))
```

# How Many Numbers Are Smaller Than the Current Number

```python
class Solution:
  def smallerNumbersThanCurrent(self, nums: list[int]) -> list[int]:
    result = []
    for i in range(len(nums)):
      count = 0
      for j in range(len(nums)):
        if j != i and nums[j] < nums[i]:
          count += 1
      result.append(count)
    return result

if __name__ == '__main__':
  sol = Solution()
  input = [8,1,2,2,3]
  print(sol.smallerNumbersThanCurrent(input))
```

# Subtract the Product and Sum of Digits of an Integer

```python
class Solution:
  def subtractProductAndSum(self, n: int) -> int:
    prod = 1
    sum = 0
    for i in str(n):
      prod *= int(i)
      sum += int(i)
    return prod - sum

if __name__ == '__main__':
  sol = Solution()
  input = 234
  print(sol.subtractProductAndSum(input))
```

# XOR Operation in an Array

```python
class Solution:
  def xorOperation(self, n: int, start: int) -> int:
    result = 0
    for i in range(n):
      result ^= start + 2 * i
    return result

if __name__ == '__main__':
  sol = Solution()
  input = 5
  input2 = 0
  print(sol.xorOperation(input, input2))
```
